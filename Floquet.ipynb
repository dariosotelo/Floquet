```julia
using Plots

function ecuacion_de_Hill(x, n, K)
    return x^n / (K^n + x^n)
end

# Valores de x para la gráfica
x_values = 0:0.1:10

# Valores de n y K para probar diferentes curvas
valores_de_n = [1, 2, 3]
valores_de_K = [1, 2, 3]

# Crear el gráfico para diferentes valores de n y K
plot(title="Ecuación de Hill", xlabel="Estímulo (x)", ylabel="Respuesta (y)", legend=:topright)

for n in valores_de_n
    for K in valores_de_K
        y_values = [ecuacion_de_Hill(x, n, K) for x in x_values]
        plot!(x_values, y_values, label="n=$n, K=$K")
    end
end

display(plot!())

```

```julia
using Plots

function ecuaciones(t, x, y)
    dy_dt = -sin(t)^2 * x
    dx_dt = y
    return dx_dt, dy_dt
end

function metodo_de_euler(x0, y0, paso, num_pasos)
    x = x0
    y = y0

    t_values = zeros(num_pasos)
    x_values = zeros(num_pasos)
    y_values = zeros(num_pasos)

    for i in 1:num_pasos
        t_values[i] = i * paso
        x_values[i] = x
        y_values[i] = y

        dx_dt, dy_dt = ecuaciones(t_values[i], x, y)
        x += paso * dx_dt
        y += paso * dy_dt
    end

    return t_values, x_values, y_values
end

# Parámetros iniciales
x0 = 0.0
y0 = 1.0
paso = 0.01
num_pasos = 1000

# Resolver el sistema de ecuaciones usando el método de Euler
t, x, y = metodo_de_euler(x0, y0, paso, num_pasos)

# Graficar la solución
plot(t, x, xlabel="Tiempo", ylabel="x(t)", label="x(t)", legend=:bottomright)
plot!(t, y, label="y(t)")
```

```julia
using Plots

function ecuaciones(t, x, y)
    dy_dt = -sin(t)^2 * x
    dx_dt = y
    return dx_dt, dy_dt
end

function metodo_de_euler(x0, y0, paso, num_pasos)
    x = x0
    y = y0

    x_values = zeros(num_pasos)
    y_values = zeros(num_pasos)

    for i in 1:num_pasos
        x_values[i] = x
        y_values[i] = y

        dx_dt, dy_dt = ecuaciones(i * paso, x, y)
        x += paso * dx_dt
        y += paso * dy_dt
    end

    return x_values, y_values
end

# Parámetros iniciales para varias trayectorias
num_trayectorias = 5
x0_values = range(-2.0, stop=2.0, length=num_trayectorias)
y0_values = range(-2.0, stop=2.0, length=num_trayectorias)

# Resolver el sistema de ecuaciones para diferentes condiciones iniciales
plot(xlabel="x", ylabel="y", legend=:bottomright)

for i in 1:num_trayectorias
    x0 = x0_values[i]
    y0 = y0_values[i]

    x, y = metodo_de_euler(x0, y0, 0.01, 10000)
    plot!(x, y, label="x0=$x0, y0=$y0")
end

display(plot!())
```
